# 1. Treating the problem as a regression problem

```python
#Since small hangups past  7,5,3 days are only used to check if there is a spike, we can drop the 3 columns
#We can also drop the timestamp columns as we keep the sequence by nto shuffling the dataset
dataset = res2.drop(['Input_Time_Window_Start', 'Input_Time_Window_End', 
                     'Target_Time_Window_Start', 'Target_Time_Window_End',
                     'Small_HangUp_Past7days', 'Small_HangUp_Past5days', 'Small_HangUp_Past3days', 'NoMajorDown'], axis=1)
print("The features we have included in this training are:\n\n", dataset.columns.to_list())

target = np.array(res2['NoMajorDown'])
#normalize the data to enable the model to learn better
array_dataset = np.array(dataset)
scaler = MinMaxScaler(feature_range = (0,1))
df = scaler.fit_transform(array_dataset)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, target, test_size=0.2, random_state=42)
```

```python
regressors = {"RandomForestRegressor": RandomForestRegressor(),
             "DecisionTreeRegressor": DecisionTreeRegressor(),
             "SGDRegressor": SGDRegressor(),
             "XGBRegressor": XGBRegressor()}

regressor_random_param = {"RandomForestRegressor": {"n_estimators": [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
                                                   "max_depth": [int(x) for x in np.linspace(10, 110, 11)],
                                                   "min_samples_split": [2, 5, 10],
                                                   "min_samples_leaf": [1, 2, 4]},
               "SGDRegressor": {"max_iter": [int(x) for x in np.linspace(100,500,5)]},
               "DecisionTreeRegressor": {"max_depth": [int(x) for x in np.linspace(10, 110, 11)],
                                         "min_samples_split": [2, 5, 10],
                                         "min_samples_leaf": [1, 2, 4]},
               "XGBRegressor": {"subsample": [0.3, 0.5, 0.7, 1],
                                "n_estimators": [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
                                "eta": [0.3, 0.5, 0.7, 1],
                                "max_depth": [int(x) for x in np.linspace(10, 110, 11)]}}
```

```python
#Helper class to return the different models training score into a dataframe
class RandomSearchHelper():
    def __init__(self, model_dict, param_dict):
        self.models = model_dict
        self.params = param_dict
        self.rs_result = {}
    
    def fit(self, x_train, y_train):
        for key in self.models.keys():
            print("Performing Random Search using {}".format(key))
            model = self.models[key]
            param = self.params[key]
            
            #since this is a regression problem, let the scoring metric be mean squared error
            rs = RandomizedSearchCV(model, param, cv=3, n_jobs=3, return_train_score=True)
            rs.fit(x_train, y_train)
            self.rs_result[key] = rs
    
    def scoring_summary(self):
        def row(key, params, scores):
            result = {"estimator": key,
                      "min_score": min(scores),
                      "max_score": max(scores),
                      "mean_score": np.mean(scores),
                      "std_score": np.std(scores)}
            return pd.Series({**result, **params})
        
        rows = []
        for k in self.rs_result:
            optimal_param = self.rs_result[k].cv_results_['params']
            
            scores = [] 
            for i in range(self.rs_result[k].cv): #returns the scores for every cross validation
                key = "split{}_test_score".format(i)
                r = self.rs_result[k].cv_results_[key] 
                scores.append(r.reshape(len(optimal_param),1))
            
            all_scores = np.hstack(scores)
            #return the result of randomized search into a dataframe
            for p, s in zip(optimal_param,all_scores):
                rows.append(row(k, p, s)) #each element in the list represent one search
            
        dataframe = pd.concat(rows, axis=1).T
        return dataframe
            
```

```python
randomsearch1 = RandomSearchHelper(regressors, regressor_random_param)
randomsearch1.fit(x_train, y_train)
```

```python
#since this is a regresssion, we cant evaluate performance using confusion matrtix and confusion report
for key, value in randomsearch1.rs_result.items():
    print(f"Evaluating with {key}, score from training: {value.best_score_}")
    model = value.best_estimator_
    history = model.predict(x_test)
    predictions = np.round(history, 0)
    
    acc_array = np.abs(predictions) - y_test #everything with numbers represent wrong predictions
    non_zero = np.count_nonzero(acc_array)
    print("Accuracy of prediction: ", (len(acc_array)- non_zero)/len(acc_array))
    
    print(predictions, y_test, "\n") 
```

# 2. Data over sampling to balance positive and negative data
##  We now treat this problem as a classification problem

```python
#ratio of alarm count to nomajordown pay attn to 1:1 ratio, alarms with many:1 ratio can be ignored
#balance target data
# SMOTE (synthetic minortiy oversampling technique)
# binary crossentropy as loss function when compiling

dataset = res2.copy()
dataset['MajorDown'] = [1 if ele != 0 else 0 for ele in res2['NoMajorDown']]
dataset['MajorDown'].value_counts() # we can see that there is an inbalance in dataset
data = dataset.drop(['Input_Time_Window_Start', 'Input_Time_Window_End', 
                     'Target_Time_Window_Start', 'Target_Time_Window_End',
                     'Small_HangUp_Past7days', 'Small_HangUp_Past5days', 'Small_HangUp_Past3days', 
                     'NoMajorDown', 'MajorDown'], axis=1)
target = dataset['MajorDown']
x_train, x_test, y_train, y_test = train_test_split(df, target, test_size=0.2, random_state=42)
y_train.value_counts() #imbalance in target values
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
x_resampled, y_resampled = sm.fit_resample(x_train, y_train)
np.bincount(y_resampled) #now there are 73 positve and 73 negative classes
```

```python
classifiers = {"RandomForestClassifier": RandomForestClassifier(),
               "SVC": SVC(),
               "DecisionTreeClassifier": DecisionTreeClassifier(),
               "XGBClassifier": XGBClassifier()}

classifier_random_param = {"RandomForestClassifier": {"n_estimators": [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
                                                       "max_depth": [int(x) for x in np.linspace(10, 110, 11)],
                                                       "min_samples_split": [2, 5, 10],
                                                       "min_samples_leaf": [1, 2, 4]},
                           "SVC": {'kernel':['linear', 'rbf', 'poly'],
                                   'C': [0.001, 0.01, 0.1, 1, 10], 
                                   'gamma' : [0.001, 0.01, 0.1, 1]},
               "DecisionTreeClassifier": {"max_depth": [int(x) for x in np.linspace(10, 110, 11)],
                                         "min_samples_split": [2, 5, 10],
                                         "min_samples_leaf": [1, 2, 4]},
               "XGBClassifier": {"subsample": [0.3, 0.5, 0.7, 1],
                                "n_estimators": [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
                                "eta": [0.3, 0.5, 0.7, 1],
                                "max_depth": [int(x) for x in np.linspace(10, 110, 11)]}}
```

```python
def fit(x_train, y_train, classifiers, classifier_param, cv):
    rs_search_result = {}
    for key, value in classifiers.items():
        print("Performing Randomized Search on {}".format(key))
        model = classifiers[key]
        param = classifier_param[key]
        
        rs = RandomizedSearchCV(model, param, cv=cv, n_jobs=-1, scoring='f1') #since this is a classification task, we can use f1 as scoring
        rs.fit(x_train, y_train)
        rs_search_result[key] = rs
    return rs_search_result
```

```python
result = fit(x_resampled, y_resampled, classifiers, classifier_random_param, 3) 
```

```python
for key, value in result.items():
    print(f"Evaluating {key} model with score {value.best_score_} with test data ")
    model = value.best_estimator_
    pred = model.predict(x_test)
    
    print(classification_report(y_test.values, pred, target_names=['NoMajorDown', 'MajorDown']))
    print(confusion_matrix(y_test.values, pred))
```

```python
#try import other alarms to make the alarms more relevant
#change setut passed, alarm19count, alarm90count, alarm3count

class Pipeline():
    def __init__(self, data, target, classifiers, params):
        self.data = data
        self.target = target
        self.classifiers = classifiers
        self.params = params
    
    def split_test_train(self, data, target, train_size):
        training_data_len = round(train_size*len(data))
        x_train = data[:training_data_len]
        y_train = target[:training_data_len]
        x_test = data[training_data_len:]
        y_test = target[training_data_len:]
        return x_train, y_train, x_test, y_test
        
    def balance_data(self, x_data, y_data):
        sm = SMOTE(random_state=42)
        x_resampled, y_resampled = sm.fit_resample(x_data, y_data)
        return x_resampled, y_resampled
        
    def fit(self, x_train, y_train, cv):
        model = self.classifiers
        param = self.params
        rs = RandomizedSearchCV(model, param, cv=cv, n_jobs=-1, scoring='f1') #since this is a classification task, we can use f1 as scoring
        rs.fit(x_train, y_train)
        return rs
    
    def evaluate_model(self, result, x_test, y_test):
        print(f"Evaluating {result.best_estimator_} model with score {result.best_score_} with test data ")
        model = result.best_estimator_
        pred = model.predict(x_test)
        
        print(confusion_matrix(y_test.values, pred))
        
    def pipeline(self, cv=3):
        params = self.params
        x_train, y_train, x_test, y_test = self.split_test_train(self.data, self.target, 0.8)
        x_balanced, y_balanced = self.balance_data(x_train, y_train)
        rs = self.fit(x_balanced,y_balanced,cv)
        self.evaluate_model(rs, x_test, y_test)
        return rs
```

```python
### remove columns with less relevance for each models ###

randomforest_data = data.drop(['Change Setup Passed', 'Change Setup Reset', 
                               'Machine Failure Passed', 'Alarm19Count', 'Alarm90Count', 'Alarm3Count',
                               'Alarm5Count'], axis=1)

decisiontree_data = data.drop(['Change Setup Passed', 'Change Setup Reset', 'Change Setup Failed', 'Change Setup', 
                               'HangUPCount', 'Alarm19Count', 'Alarm90Count', 'Alarm3Count',
                               'Alarm5Count', 'Alarm10Count', 'Alarm2Count'], axis=1)

XGB_data = data.drop(['Days since last maintenance', 'Change Setup Reset', 'Change Setup Passed', 'Change Setup', 
                      'Alarm19Count', 'Alarm90Count', 'Alarm3Count', 
                      'Alarm2Count', 'Alarm5Count','Proportion of 5 days hangup over 7 days'], axis=1)
```

```python
### pipelines for each tree based models ###

randomforest_params = {"n_estimators": [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
                        "max_depth": [int(x) for x in np.linspace(10, 110, 11)],
                        "min_samples_split": [2, 5, 10],
                        "min_samples_leaf": [1, 2, 4]}

decisiontree_params = {"max_depth": [int(x) for x in np.linspace(10, 110, 11)],
                       "min_samples_split": [2, 5, 10],
                       "min_samples_leaf": [1, 2, 4]}

XGB_params = {"subsample": [0.3, 0.5, 0.7, 1],
            "n_estimators": [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
            "eta": [0.3, 0.5, 0.7, 1],
            "max_depth": [int(x) for x in np.linspace(10, 110, 11)]}

SVC_params = {'kernel':['linear', 'rbf', 'poly'],
               'C': [0.001, 0.01, 0.1, 1, 10], 
               'gamma' : [0.001, 0.01, 0.1, 1]}
```

```python
### pipeline for randomforest ###
randomforest_pipeline = Pipeline(randomforest_data, target, RandomForestClassifier(random_state=42), randomforest_params)
rf = randomforest_pipeline.pipeline()

### pipeline for decisiontree ###
decisiontree_pipeline = Pipeline(decisiontree_data, target, DecisionTreeClassifier(random_state=42), decisiontree_params)
dt = decisiontree_pipeline.pipeline()

### pipeline for XGB ###
XGB_pipeline = Pipeline(XGB_data, target, XGBClassifier(random_state=42), XGB_params)
xgb = XGB_pipeline.pipeline()
```

# 4. Remove columns that are highly correlated and with low variance

```python
import seaborn as sns 
import matplotlib.pyplot as plt

#plot correlation matrix for each columns
corr = data.corr()
plt.figure(figsize=(16,9))
sns.heatmap(corr, annot=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in dataset.columns:
                    del dataset[colname] # deleting the column from the dataset
    return dataset
```

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
corr_data = data.copy()
new_corr = scaler.fit_transform(corr_data)

x = new_corr
v_threshold = VarianceThreshold(threshold=0.05)
v_threshold.fit(x)
v_threshold.get_support() #boolean array on which columns have variance larger than 0.05
removed = [column for column in corr_data.columns if column not in corr_data.columns[v_threshold.get_support()]]
print('Columns with low variance: ', removed)

new_mat = corr_data.drop(removed,axis=1)
corr = correlation(new_mat, 0.85)
removed = [column for column in corr_data.columns if column not in corr.columns]
print('Columns with low variance and high correlation: ', removed)
```

```python
### minmaxscale the data to make the model learn better ###
scaled_data = scaler.fit_transform(corr)

### pipeline for randomforest ###
randomforest_pipeline = Pipeline(scaled_data, target, RandomForestClassifier(random_state=42), randomforest_params)
rf = randomforest_pipeline.pipeline()

### pipeline for decisiontree ###
decisiontree_pipeline = Pipeline(scaled_data, target, DecisionTreeClassifier(random_state=42), decisiontree_params)
dt = decisiontree_pipeline.pipeline()

### pipeline for XGB ###
XGB_pipeline = Pipeline(scaled_data, target, XGBClassifier(random_state=42), XGB_params)
xgb = XGB_pipeline.pipeline()

### pipeline for SVC ###
SVC_pipeline = Pipeline(scaled_data, target, SVC(random_state=42), SVC_params)
svc = SVC_pipeline.pipeline()
```

```python
from sklearn.base import clone
from sklearn.pipeline import make_pipeline  
from sklearn.preprocessing import MinMaxScaler #model learns better from scaled data
from sklearn.feature_selection import SelectFromModel  

training_data_len = round(0.9*len(data))
x_train, y_train = data[:training_data_len], target[:training_data_len]
x_test, y_test = data[training_data_len:], target[training_data_len:]

rf = RandomForestClassifier(random_state=42, n_jobs=-1)
dt = DecisionTreeClassifier(random_state=42)
xgb = XGBClassifier(random_state=42, n_jobs=-1)
svc = SVC(random_state=42)

models = [rf, dt, xgb, svc]
params = [randomforest_params, decisiontree_params, XGB_params, SVC_params]

for model, param in  zip(models,params):
    pipe = make_pipeline(MinMaxScaler(), SelectFromModel(model), clone(model))
    rs = RandomizedSearchCV(model, param, scoring='f1', n_jobs=-1)
    rs.fit(x_train, y_train)
    preds = rs.best_estimator_.predict(x_test)
    print(f"Confusion Matrix for {str(model)}")
    print(confusion_matrix(y_test, preds))
```
